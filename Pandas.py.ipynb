```python
#Task for excel and csv
import pandas as pd
import numpy as np
salary = np.array([101000, 100000, 102000, 101000, 100000,  101000,  100000, 100000, 101000,  100000,  101000, 100000,])
LOP =  np.array([2000, 3000, 3000,  3000,  3000,  4000, 4000,  4000,  2000,  1000,  11000,  2000])
allowance = np.array([20000, 30000, 30000, 30000, 30000  ,9000,  4000,  9000,  9000,  1000,  11000,  20000])
print('Checking size for all 3',salary.size,LOP.size,allowance.size)
Total_salary = salary - LOP + allowance
print('')
print('Total take home',Total_salary)
ctpf = {"Salary":salary, "Loss_of_Pay":LOP, "Allowance":allowance, "Take_Home":Total_salary}
new_value = pd.DataFrame(ctpf)
print(new_value)
```

```python
import pandas as pd
import numpy as np
r_sal = pd.read_excel("salary.xlsx")
#print(r_sal)
idn = r_sal["IDNumber"].to_numpy().flatten()
sal = r_sal["Salary"].to_numpy().flatten()
r_da = r_sal["DA"].to_numpy().flatten()
r_tax = r_sal["Tax"].to_numpy().flatten()
r_take_home = sal + r_da - r_tax
frame = {"IDNumber":idn,"Salary":sal,"DA":r_da,"Tax":r_tax,"Take_Home":r_take_home}
new = pd.DataFrame(frame)
print(new)
new.to_excel("Updated_sal.xlsx",index=False)
```
